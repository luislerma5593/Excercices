```python
import os
from skimage.metrics import structural_similarity as compare_ssim
import cv2
import numpy as np
from tqdm import tqdm
import matplotlib.pyplot as plt
import pandas as pd
```

### Comparison function

```python
def ComparePictures(imagenA,imagenB):
    global filesAuto
    imageA = cv2.imread(imagenA)
    imageB = cv2.imread(imagenB)

    grayA = cv2.cvtColor(imageA, cv2.COLOR_RGB2GRAY)
    grayB = cv2.cvtColor(imageB, cv2.COLOR_RGB2GRAY)

    (score, diff) = compare_ssim(grayA, grayB, full=True)
    diff = (diff * 255).astype("uint8")

    #print("SSIM: {}".format(score))
    return score
```

### Dataset

```python
in_path = r'C:\Users\LLerma\Documents\AI\IoConveyorCharlotte\Images\Dataset\Set\DataSet\\'

for root, dirs, files in os.walk(in_path, topdown=False):
   ''

len_dirs = len(dirs)
arr_files = [[] for x in range(len_dirs)]
arr_scores_dataset = [[] for x in range(len_dirs)]

for i in range(len_dirs):
    for filename in os.listdir(in_path+dirs[i]):
        arr_files[i].append(os.path.join(in_path,dirs[i],filename))

for directory in range(len_dirs):

    steps = 10
    start = 0
    finish = steps
    windows = int(len(arr_files[directory]) / steps)

    for window in tqdm(range(windows)):

        array = []
        control_image = arr_files[directory][start]

        for i in range(start , finish-1):
            current_image = (arr_files[directory][i+1])
            score = ComparePictures(control_image, current_image)
            array.append(score)
        arr_scores_dataset[directory].append(round(np.mean(array),5))
        #print(control_image)
        #print(current_image)
        start+=steps
        finish+=steps

'''for scrap_class in range(len_dirs):
    print(dirs[scrap_class],'\n', arr_scores_dataset[scrap_class])
    print('\n')
    

np.mean(arr_scores_dataset, axis=1)'''
```
